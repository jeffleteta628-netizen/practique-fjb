## Exercice 1 : Lecture de code
### Que fait ce code ? Compléter les commentaires

```python
import matplotlib.pyplot as plt

# Données
pays = ['Sénégal', 'Mali', 'Guinée']
population = [17, 22, 13]  # en millions

# Question 1 : Que fait cette ligne ?
fig, ax = plt.subplots()
# Réponse : Cette ligne crée une figure (fig) et un axe (ax) vides.
# fig représente la fenêtre du graphique entier,
# ax est la zone dans laquelle les données vont être tracées.

# Question 2 : Que va afficher cette ligne ?
ax.bar(pays, population, color=['green', 'yellow', 'red'])
# Réponse : Cette ligne crée un graphique en barres (histogramme).
# Chaque pays est représenté par une barre colorée :
# Sénégal en vert, Mali en jaune, Guinée en rouge.
# La hauteur de chaque barre correspond à la population en millions.

# Question 3 : Quel est le rôle de ax.set_ylabel ?
ax.set_ylabel('Population (millions)')
# Réponse : Cette fonction définit le titre (l'étiquette) de l'axe Y.
# Ici elle indique que l'axe vertical représente la population exprimée en millions,
# ce qui permet au lecteur de comprendre l'unité utilisée.

ax.set_title('Population de trois pays ouest-africains')
plt.show()
```

## Exercice 2 : Choix du bon graphique

### Réponses :

| Situation | Type de graphique |
|-----------|-------------------|
| 1. Évolution du CA sur 12 mois | **B. Line plot (courbe)** |
| 2. Comparer les salaires Hommes/Femmes | **E. Bar chart groupé** |
| 3. Voir la répartition des âges des clients | **D. Histogramme** |
| 4. Relation entre prix et superficie d'un logement | **C. Scatter plot** |
| 5. Part de marché de 3 concurrents | **F. Camembert** |
| 6. Détecter des valeurs aberrantes dans les salaires | **A. Boxplot** |

**Réponses résumées :** 1-B , 2-E , 3-D , 4-C , 5-F , 6-A

### Justifications :
- **1 → B (Line plot)** : une courbe est idéale pour représenter une évolution dans le temps (tendance sur 12 mois).
- **2 → E (Bar chart groupé)** : les barres groupées permettent de comparer deux catégories (H/F) côte à côte.
- **3 → D (Histogramme)** : l'histogramme montre comment une variable numérique (âge) se distribue dans une population.
- **4 → C (Scatter plot)** : le nuage de points révèle la corrélation entre deux variables numériques continues.
- **5 → F (Camembert)** : le camembert (pie chart) illustre bien les proportions quand on a peu de catégories.
- **6 → A (Boxplot)** : la boîte à moustaches est spécialement conçue pour détecter les outliers et la dispersion.

## Exercice 3 : Corriger un mauvais graphique

### Questions :

**1. Quel est le problème principal de ce graphique ?**

Le problème principal est l'utilisation d'un **camembert (pie chart) avec trop de catégories** (10 produits). Avec autant de parts, le graphique devient illisible : les petites parts (Produit G, H, I, J qui font moins de 10% chacune) sont pratiquement invisibles, et il est impossible de comparer visuellement des parts proches en taille. De plus, le titre "Ventes" est trop vague.

**2. Proposer un meilleur type de graphique :**

Un **graphique en barres horizontales (bar chart horizontal)** est plus adapté car :
- Il permet de lire clairement les valeurs de chaque produit
- On peut facilement trier du plus grand au plus petit
- Les étiquettes de catégories sont lisibles horizontalement
- Les différences entre valeurs proches (ex: 50, 45) restent visibles

**3. Code amélioré :**

```python
# CODE AMÉLIORÉ — Exercice 3
import matplotlib.pyplot as plt

categories = ['Produit A', 'Produit B', 'Produit C', 'Produit D', 'Produit E',
              'Produit F', 'Produit G', 'Produit H', 'Produit I', 'Produit J']
ventes = [150, 220, 180, 300, 250, 50, 45, 30, 25, 20]

# Trier les données du plus grand au plus petit pour une meilleure lisibilité
donnees = sorted(zip(ventes, categories), reverse=True)
ventes_tries, categories_tries = zip(*donnees)

fig, ax = plt.subplots(figsize=(10, 6))

# Barres horizontales avec une couleur qui varie selon la valeur
barres = ax.barh(categories_tries, ventes_tries, color='steelblue', edgecolor='white')

# Ajouter les valeurs sur chaque barre
for barre, valeur in zip(barres, ventes_tries):
    ax.text(barre.get_width() + 3, barre.get_y() + barre.get_height() / 2,
            str(valeur), va='center', fontsize=10)

# Titres et étiquettes clairs
ax.set_xlabel('Nombre de ventes')
ax.set_title('Ventes par produit (classées du plus au moins vendu)', fontsize=13, fontweight='bold')
ax.set_xlim(0, 340)  # un peu d'espace pour les étiquettes de valeur

plt.tight_layout()
plt.show()

print("Le bar chart horizontal est beaucoup plus lisible que le camembert avec 10 catégories.")
```

## Exercice 4 : Créer son premier graphique Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Charger les données
mpg = sns.load_dataset('mpg')
print(mpg.head())
print("\nColonnes disponibles :", mpg.columns.tolist())
print("\nNombre de lignes :", len(mpg))
print("Valeurs manquantes :\n", mpg.isnull().sum())
```

```python
# Question 1 : Scatter plot weight vs mpg
fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(data=mpg, x='weight', y='mpg', ax=ax, alpha=0.6, color='royalblue')
ax.set_title('Relation entre le poids et la consommation des voitures', fontsize=13)
ax.set_xlabel('Poids du véhicule (livres)')
ax.set_ylabel('Consommation (miles par gallon)')
plt.tight_layout()
plt.show()

print("""
Constat Question 1 :
On observe une relation négative (corrélation inverse) entre le poids et la consommation.
Plus une voiture est lourde, moins elle est économe en carburant (mpg faible).
Les points forment un nuage qui descend de gauche à droite, ce qui confirme cette tendance.
""")
```

```python
# Question 2 : Scatter plot avec hue='origin'
fig, ax = plt.subplots(figsize=(9, 6))
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin',
                palette={'usa': '#e74c3c', 'europe': '#2ecc71', 'japan': '#3498db'},
                alpha=0.7, ax=ax)
ax.set_title('Poids vs Consommation selon l\'origine du véhicule', fontsize=13)
ax.set_xlabel('Poids du véhicule (livres)')
ax.set_ylabel('Consommation (miles par gallon)')
ax.legend(title='Origine')
plt.tight_layout()
plt.show()

# Vérification numérique
print("Consommation moyenne par origine :")
print(mpg.groupby('origin')['mpg'].mean().round(2))
print("""
Constat Question 2 :
Les voitures japonaises (bleu) sont les plus économes : elles ont les valeurs mpg les plus élevées
et sont concentrées dans les poids légers.
Les voitures américaines (rouge) sont les plus lourdes et les moins économes.
Les voitures européennes (vert) se situent entre les deux.
""")
```

```python
# Question 3 : Boxplot consommation selon le nombre de cylindres
fig, ax = plt.subplots(figsize=(9, 5))

# Supprimer les lignes avec des valeurs manquantes pour mpg
mpg_clean = mpg.dropna(subset=['mpg'])

sns.boxplot(data=mpg_clean, x='cylinders', y='mpg',
            palette='Blues', ax=ax)
ax.set_title('Distribution de la consommation selon le nombre de cylindres', fontsize=13)
ax.set_xlabel('Nombre de cylindres')
ax.set_ylabel('Consommation (miles par gallon)')
plt.tight_layout()
plt.show()

print("Médiane de consommation par nombre de cylindres :")
print(mpg_clean.groupby('cylinders')['mpg'].median().round(2))
print("""
Constat Question 3 :
Le boxplot confirme clairement que plus une voiture a de cylindres, plus elle consomme (mpg plus bas).
Les voitures à 4 cylindres sont nettement plus économes que celles à 6 ou 8 cylindres.
Les boîtes à 8 cylindres montrent peu de dispersion (valeurs groupées autour de 15 mpg),
tandis que les 4 cylindres ont une distribution plus large.
""")
```

## Exercice 5 : Analyse de graphiques trompeurs

**Situation décrite :** Un graphique en barres montre les ventes du Produit X : 100 en 2024, 102 en 2025. L'axe Y commence à 95 et va jusqu'à 103. La barre de 2025 est deux fois plus haute que celle de 2024.

---

**1. Quelle est l'augmentation réelle en pourcentage ?**

Augmentation réelle = ((102 - 100) / 100) × 100 = **2%**

C'est une très faible progression de seulement 2 unités.

---

**2. Pourquoi le graphique donne-t-il une impression trompeuse ?**

Le graphique est trompeur à cause de l'**axe Y tronqué** : il commence à 95 au lieu de 0. Cette technique visuelle exagère visuellement la différence entre les deux barres. Comme l'axe ne part pas de zéro, la barre de 2025 (102) paraît deux fois plus haute que celle de 2024 (100), alors que la différence réelle n'est que de 2 unités sur une base de 100. Cette manipulation peut faire croire à une croissance spectaculaire alors qu'elle est minime.

---

**3. Comment corriger ce graphique ?**

- Faire commencer l'axe Y à **0** (règle de base pour les graphiques en barres)
- Ou, si on veut zoomer sur la variation, utiliser un **line plot** (courbe) avec une note expliquant que l'axe est tronqué
- Afficher le **pourcentage d'augmentation** directement sur le graphique pour que le lecteur comprenne l'ordre de grandeur réel
- Ajouter un titre honnête comme "Légère augmentation de 2%" plutôt qu'un titre neutre qui laisse croire à une forte hausse

```python
# Démonstration visuelle de l'effet trompeur vs graphique correct
import matplotlib.pyplot as plt

annees = ['2024', '2025']
ventes = [100, 102]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Graphique trompeur (axe Y tronqué)
ax1.bar(annees, ventes, color=['#3498db', '#e74c3c'])
ax1.set_ylim(95, 103)
ax1.set_title('Version trompeuse\n(axe Y commence à 95)', color='red', fontsize=11)
ax1.set_ylabel('Ventes')

# Graphique correct (axe Y commence à 0)
ax2.bar(annees, ventes, color=['#3498db', '#2ecc71'])
ax2.set_ylim(0, 120)
ax2.set_title('Version correcte\n(axe Y commence à 0 — augmentation réelle = 2%)', color='green', fontsize=11)
ax2.set_ylabel('Ventes')

# Ajouter les valeurs sur les barres
for ax in [ax1, ax2]:
    for i, (annee, valeur) in enumerate(zip(annees, ventes)):
        ax.text(i, valeur + 0.2, str(valeur), ha='center', fontweight='bold')

plt.suptitle('Impact de l\'axe tronqué sur la perception du graphique', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```
